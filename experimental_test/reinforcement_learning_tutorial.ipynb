## 1. 強化学習の枠組み
### 1.1 強化学習で扱う問題

　強化学習は, `エージェント`と呼ばれる行動主体が, `環境`の中で試行錯誤を繰り返しながら, 各`状態`における最適な`行動`を選択する`方策`を学習するアルゴリズムである. 例えば`エージェント`は自動走行する車などが当てはまり, `環境`はエージェントが試行錯誤する場として道路が当てはまる. 
  強化学習では`方策`の良し悪しを測る指標として`報酬`が定義される. エージェントは行動の良し悪しによって, それに見合う`報酬`を獲得する. 一定期間において獲得可能な`報酬`を最大(最小)化するような方策が最適な方策ということになる.
  
　具体的な試行錯誤としては, 以下の図のように初期状態①が最初に与えられ, エージェントはその状態における行動を方策を元に決定する②. さらに, エージェントが実行した行動によって環境は次の状態に遷移する③. 最後に遷移した状態の良し悪しによって,
 (一つ前の状態(ここでは初期状態), 行動, 遷移後の状態)に対し報酬が与えられる. 報酬はコストとして考えてもよく, その場合はコストが最小となる方策を求める問題となる.



具体的な定義:

行動: $ a \in Action\_space $  
状態: $ s \in State\_space $  
報酬関数: $ c(s): s \to r \in R $  
方策: $ \pi(a, s): a, s \to [0, 1] $


### 1.2 強化学習にグルーピングされるアルゴリズム

強化学習の概念は非常に抽象的であるため, あらゆる問題が強化学習の枠組みに捉えられる.  
その中でも以下のアルゴリズムは, 強化学習の特殊なケースとして考えることができる.

#### 1 バンディットアルゴリズム(バンディット問題)

バンディット問題では, エージェントが実行した`行動`が次の状態に反映されないような問題を扱う. これは例えば, ユーザにとって最適なweb広告を表示する問題がそれに当たる. この場合, エージェントは「表示する広告を選択する主体」であり, 行動は「広告を選択して表示する」こととなる.
